```python
import pandas as pd
```

```python
SAT_data_path = "Resources/AVG_SAT_NYC_2014_2015.csv"

SAT_df = pd.read_csv(SAT_data_path)
SAT_df.head()
```

```python
SAT_df.shape
```

```python
dropped_SAT = SAT_df.dropna()
dropped_SAT.head()
```

```python
dropped_SAT.shape
```

```python
named_SAT = dropped_SAT.drop(columns = ["School ID", "Building Code", "Phone Number", "Street Address", "Start Time", "End Time"])
named_SAT.head()
```

```python
named_SAT.shape
```

```python
named_SAT.dtypes
```

```python
named_SAT["Percent White"] = named_SAT["Percent White"].replace('%', '', regex=True).astype('float')
named_SAT["Percent Black"] = named_SAT["Percent Black"].replace('%', '', regex=True).astype('float')
named_SAT["Percent Asian"] = named_SAT["Percent Asian"].replace('%', '', regex=True).astype('float')
named_SAT["Percent Hispanic"] = named_SAT["Percent Hispanic"].replace('%', '', regex=True).astype('float')
named_SAT["Percent Tested"] = named_SAT["Percent Tested"].replace('%', '', regex=True).astype('float')
```

```python
named_SAT.dtypes
```

```python
clean_SAT = named_SAT
```

```python
clean_SAT.to_csv("Resources/clean_SAT.csv", index = False)
```

```python
# Rename columns for Postgres loading

sqlSAT = clean_SAT.rename(columns={"School Name": "school_name", "Borough": "borough", "City": "city", "State": "state",
                                  "Zip Code": "zip_code", "Latitude": "latitude", "Longitude": "longitude",
                                  "Student Enrollment": "student_enrollment", "Percent White": "percent_white",
                                  "Percent Black": "percent_black", "Percent Hispanic": "percent_hispanic",
                                  "Percent Asian": "percent_asian", "Average Score (SAT Math)": "avg_score_math",
                                  "Average Score (SAT Reading)": "avg_score_reading", "Average Score (SAT Writing)":
                                  "avg_score_writing", "Percent Tested": "percent_tested"})
```

```python
sqlSAT.head()
```

```python
sqlSAT.to_csv("Resources/sqlSAT.csv", index = False)
```
